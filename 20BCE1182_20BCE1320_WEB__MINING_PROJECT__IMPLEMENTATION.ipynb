```python
import csv
from tqdm import tqdm
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import networkx as nx 
G = nx.DiGraph()
```

```python
with open('contacts.csv', 'r') as f:
    data = csv.reader(f)
    i=1
    headers = next(data)
    for row in tqdm(data):
        i=i+1
        if i==201:
            break
        G.add_node(row[0]) #superhero in first column
        G.add_node(row[1]) #superhero in second column
        if G.has_edge(row[0], row[1]):
            # edge already exists, increase weight by one
            G[row[0]][row[1]]['weight'] += 1
        else:
            # add new edge with weight 1
            G.add_edge(row[0], row[1], weight = 1)
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

# **Calculating the degree centrality**

```python
deg_centrality = nx.degree_centrality(G)
print(deg_centrality)
```

# **Visualizing Degree Centrality**

```python
l=[deg_centrality[key] for key in deg_centrality]
a=np.array(l)
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(a)
plt.show()
```

# **Calculating Closeness Centrality**

```python
close_centrality = nx.closeness_centrality(G)
print(close_centrality)
```

# **Visualising Closeness Centrality**

```python
l=[close_centrality[key] for key in close_centrality]
a=np.array(l)
fig, ax = plt.subplots(figsize =(12, 7))
ax.hist(a)
plt.show()
```

# **Calculating Betweeness Centrality**

```python
bet_centrality = nx.betweenness_centrality(G, normalized = True, 
                                              endpoints = True)
print(bet_centrality)
```

# **Visualising Betweeness Centrality**

```python
l=[bet_centrality[key] for key in bet_centrality]
a=np.array(l)
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(a)
plt.show()
```

# **Calculating eigen value centrality**

```python
eigenvector_centrality = nx.eigenvector_centrality(G,200,1.0e-6)
print(eigenvector_centrality)
```

# **Highest Eigen value**

```python
l=[eigenvector_centrality[key] for key in eigenvector_centrality]
mx=max(l)
lst=[key for key in eigenvector_centrality if eigenvector_centrality[key]==mx]
print(lst)
```

```python
#highest eigen vector value

mx=max(l)
lst=[key for key in eigenvector_centrality if eigenvector_centrality[key]==mx]
print(lst)
```

# **Clustering**

Agglomerative Hierarchical Clustering(Dendogram)


```python
import pandas as pd
import networkx as nx
import seaborn as sns
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram,linkage
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("contacts.csv")
w1=df.iloc[:,0:1].values
z1=linkage(w1)
dendrogram(z1, leaf_rotation=45., leaf_font_size=12. , show_contracted=True)
plt.style.use("seaborn-whitegrid")
plt.title("Dendogram to find clusters")
plt.ylabel("Distance")
plt.figure(figsize=(100,100))
plt.show()
```
